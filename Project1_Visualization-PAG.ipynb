```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
#Set seaborn default style for matplotlib
sns.set()   

# this statement allows the visuals to render within your Jupyter Notebook
%matplotlib inline 
```

```python
# Loading in the data for all stations
stations = pd.read_csv('station_hospital_distances.csv')
```

```python
stations.head()
```

```python
# Histogram of all average daily entries for all stations
plt.hist(stations['DAILY_ENTRIES'], bins=10);
```

```python
# just top 10 stations to match to hospitals
top_stns = stations
top_stns.sort_values(by=['DAILY_ENTRIES'])
top_stns = top_stns.head(10)
# top_stns.head(11)
```

```python
# Plots all entries for all stations

# NEED HELP HERE - labelpad = '40' does not work for xticks
chart = sns.barplot(top_stns.STATION_LINE, top_stns['DAILY_ENTRIES'], ci = None);
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right');
```

```python
# This should get prettier with cleaner data

# Box plots by station
sns.boxplot('DAILY_ENTRIES', 'STATION_LINE', data=top_stns);
```

```python

#Box plot of all
sns.boxplot(data=top_stns);
```
